# Generate Ontario Holiday Features for 2013-2025

**Purpose:** Create holiday indicators for electricity demand forecasting.

**Why holidays matter:** Statutory holidays significantly impact electricity demand:
- Commercial/industrial facilities close (10-30% demand reduction)
- Residential patterns shift (different usage timing)
- Long weekends create travel patterns affecting urban demand

**Ontario Statutory Holidays (9 total):**
1. New Year's Day (January 1)
2. Family Day (3rd Monday in February)
3. Good Friday (variable, based on Easter)
4. Victoria Day (Monday before May 25)
5. Canada Day (July 1)
6. Labour Day (1st Monday in September)
7. Thanksgiving (2nd Monday in October)
8. Christmas Day (December 25)
9. Boxing Day (December 26) - Ontario only!

**Time Range:** June 2013 - November 2025

**Method:** Python `holidays` package with Ontario-specific calendar

```python
# Install holidays package if not already installed
import sys
!{sys.executable} -m pip install holidays --quiet

# Import libraries
import pandas as pd
import numpy as np
import holidays
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

print("✓ Libraries imported successfully")
print(f"✓ holidays package version: {holidays.__version__}")
```

```python
# Generate Ontario holiday calendar for 2013-2025
print("Generating Ontario statutory holidays for 2013-2025...")

# Create Ontario-specific holiday calendar
on_holidays = holidays.Canada(subdiv='ON', years=range(2013, 2026))

print(f"✓ Generated {len(on_holidays)} holiday dates\n")

# Convert to DataFrame for easier viewing
holiday_list = []
for date, name in sorted(on_holidays.items()):
    holiday_list.append({'Date': date, 'Holiday_Name': name})

df_holidays_list = pd.DataFrame(holiday_list)

# Convert Date to datetime for filtering
df_holidays_list['Date'] = pd.to_datetime(df_holidays_list['Date'])

# Filter to our timeframe (June 2013 onwards)
df_holidays_list = df_holidays_list[df_holidays_list['Date'] >= '2013-06-01'].copy()

print(f"Holidays in our timeframe (June 2013 - Dec 2025): {len(df_holidays_list)}")
print(f"\nFirst 10 holidays:")
print(df_holidays_list.head(10))

print(f"\nLast 10 holidays:")
print(df_holidays_list.tail(10))
```

```python
# Save the holiday list
output_path = "../../02_Datasets/processed/ontario_holidays_2013_2025.csv"
df_holidays_list.to_csv(output_path, index=False)

print(f"✓ Holiday data saved to: {output_path}")
print(f"\nFinal holiday dataset:")
print(f"  Shape: {df_holidays_list.shape}")
print(f"  Date range: {df_holidays_list['Date'].min()} to {df_holidays_list['Date'].max()}")
print(f"  Total holidays: {len(df_holidays_list)}")
print(f"  Memory usage: {df_holidays_list.memory_usage(deep=True).sum() / 1024:.2f} KB")

# Count holidays by name
print(f"\nHoliday frequency (2013-2025):")
print(df_holidays_list['Holiday_Name'].value_counts().sort_values(ascending=False))

print(f"\n{'='*70}")
print(f"ONTARIO HOLIDAY FEATURES COMPLETE! ✅")
print(f"{'='*70}")
```
